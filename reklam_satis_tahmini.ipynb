```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("C:/Users/usnis/OneDrive/Masaüstü/Python_Workspace/reklam.csv")
```

```python
data.head()
```

```python
x=data.TV.values.reshape(-1,1)
```

```python
y=data.satış.values.reshape(-1,1)
```

```python
plt.scatter(x,y)
plt.plot(x,y)
plt.title("TV-Satış Grafiği")
plt.xlabel("TV")
plt.ylabel("satış")
plt.show
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=12)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
lr.fit(xtrain, ytrain)
```

```python
yhead=lr.predict(xtest)
```

```python
yhead
```

```python
ytest
```

```python
plt.scatter(x,y)
plt.plot(xtest, yhead, color="purple")
```

```python
from sklearn.metrics import mean_squared_error, r2_score

rmse=np.sqrt(mean_squared_error(ytest, yhead))
print(rmse)
```

```python
r2=r2_score(ytest,yhead)
```
