```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("C:/Users/usnis/OneDrive/Masaüstü/Python_Workspace/poly.csv")
data.head()
```

```python
data.tail()
```

```python
x=data.iloc[:,0].values.reshape(-1,1)
y=data.iloc[:,1].values
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
lr.fit(x,y)
```

```python
plt.scatter(x,y)
plt.plot(x,lr.predict(x))
```

```python
from sklearn.preprocessing import PolynomialFeatures
pr=PolynomialFeatures(degree=4)
```

```python
xpl=pr.fit_transform(x)
xpl
```

```python
lr2=LinearRegression()
lr2.fit(xpl,y)
```

```python
plt.scatter(x,y)
plt.plot(x,lr2.predict(xpl),color="purple")
```
