```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data=pd.read_csv("C:/Users/usnis/OneDrive/Masaüstü/Python_Workspace/musteri.csv")
data.head()
```

```python
x=data.iloc[:,[3,4]].values
x
```

```python
import warnings
warnings.filterwarnings('ignore')
#uyarıları görmezden gel
```

```python
from sklearn.cluster import KMeans
wcss=[]
for i in range(1,14):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
    #dirsek methodunu uygulama
```

```python
plt.plot(range(1,14),wcss)
```

```python
kmeans=KMeans(n_clusters=5)
kmeans.fit(x)

plt.scatter(x[:,0],x[:,1],c=kmeans.labels_, cmap="rainbow",label="geliir-harcama")
plt.title("clusters")
plt.xlabel("gelir")
plt.ylabel("harcama puanı")
plt.legend()
plt.show()
```

```python
plt.scatter(x[:,0],x[:,1],c=kmeans.labels_, cmap="rainbow")
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=100,c="black",label="centroid")
plt.title("clusters")
plt.xlabel("gelir")
plt.ylabel("harcama puanı")
plt.legend()
plt.show()
```

```python
manuel_value=np.array([[22,33]])
plt.scatter(x[:,0],x[:,1],c=kmeans.labels_, cmap="rainbow")
kume_no=kmeans.predict(manuel_value)
plt.scatter(manuel_value[:,0],manuel_value[:,1],s=200,c="black",marker="x")
plt.title("clusters")
plt.xlabel("gelir")
plt.ylabel("harcama puanı")
plt.legend()
plt.show()
```
